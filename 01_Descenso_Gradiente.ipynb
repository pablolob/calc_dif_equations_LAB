# 🧬 LAB 1: Escalar las Montañas Metabólicas (Descenso del Gradiente)

¡Bienvenidos al primer laboratorio! Aquí combinaremos Cálculo Diferencial y Biomedicina.

Veremos un problema de minimización en farmacología y como el algoritmo de descenso por gradiente puede ayudar para encontrar soluciones. 

**El Caso Clínico**: En la optimización de regímenes de dosificación, los farmacólogos buscan la combinación de dos fármacos que minimiza la toxicidad para el paciente. Sabemos que la toxicidad sigue este modelo:

$$T(x,y) = (x-2)^2 + 4(y-1)^2$$

Nuestro objetivo es, partiendo de unas dosis iniciales elevadas, **dejar que el algoritmo encuentre por sí solo las dosis óptimas** usando derivadas parciales: **El Descenso del Gradiente**.


---

````python
# !!! ESTA CELDA ES SOLO DE CONFIGURACIÓN.
# Ejecútala para cargar nuestras herramientas de dibujo y no le des muchas vueltas.
import numpy as np
import matplotlib.pyplot as plt

def dibujar_toxicidad(historico_x, historico_y):
    """Dibuja el mapa topográfico y los pasos del algoritmo"""
    X, Y = np.meshgrid(np.linspace(-2, 7, 60), np.linspace(-2, 5, 60))
    Z = (X - 2)**2 + 4 * (Y - 1)**2
    
    x_final = historico_x[-1]
    y_final = historico_y[-1]
    
    plt.figure(figsize=(8, 6))
    plt.contourf(X, Y, Z, levels=25, cmap='plasma')
    plt.colorbar(label='Nivel de Toxicidad')
    plt.plot(historico_x, historico_y, 'wo-', markersize=6, label='Tu recorrido')
    plt.plot(x_final, y_final, 'g*', markersize=20,
             label=f'Dosis encontrada: x={x_final:.2f}, y={y_final:.2f}')
    plt.legend()
    plt.title('Mapa de Contorno de $T(x,y) = (x-2)^2 + 4(y-1)^2$')
    plt.xlabel('Dosis fármaco X (mg/kg)')
    plt.ylabel('Dosis fármaco Y (mg/kg)')
    plt.show()
````

### 🛠️ ¡TU TURNO: A PROGRAMAR!

El descenso por gradiente, consiste en repetir un movimiento en la dirección que más puede mejorar el resultado.

La regla para dar un "paso" cuesta abajo en matemáticas es estar en la posición actual y restarle un "trozo" delgradiente (que apunta a donde más sube).

$$X_{nueva} = X_{actual} - \alpha \cdot \frac{\partial T}{\partial X}$$
$$Y_{nueva} = Y_{actual} - \alpha \cdot \frac{\partial T}{\partial Y}$$

$\alpha$ (alpha) es la **tasa de aprendizaje** (learning rate). Es "cuánto" nos atrevemos a saltar.

Sabiendo que $T(x,y) = (x-2)^2 + 4(y-1)^2$, calcula las derivadas parciales aplicando la regla de la cadena:
- La derivada parcial respecto a x es: $2(x-2)$
- La derivada parcial respecto a y es: $8(y-1)$

````python
x_actual = 5.0   # Empezamos con una dosis inicial alejada del óptimo
y_actual = 4.0   # Empezamos con una dosis inicial alejada del óptimo

tasa_aprendizaje = 0.1  # Prueba a cambiarlo luego a 0.01 o a 0.2
pasos = 20

historico_x = [x_actual]
historico_y = [y_actual]

for i in range(pasos):
    # 1. Calcula los gradientes (derivadas parciales)
    gradiente_x = 2 * (x_actual - 2)
    gradiente_y = 8 * (y_actual - 1)
    
    # 2. ¡REEMPLAZA EL 'None' POR TU FÓRMULA! 
    x_actual = None    # <--- ¡Mete aquí la fórmula del descenso!
    y_actual = None    # <--- ¡Mete aquí la fórmula del descenso!
    
    historico_x.append(x_actual)
    historico_y.append(y_actual)

# Dibujamos tu resultado mágico:
dibujar_toxicidad(historico_x, historico_y)
````
